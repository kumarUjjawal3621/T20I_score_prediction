```python
import numpy as np
import pandas as pd
from yaml import safe_load
import os
from tqdm import tqdm
```

```python
filenames = []
for file in os.listdir('kaggle_data'):
    filenames.append(os.path.join('kaggle_data',file))
```

```python
print(filenames[0:5])
type(filenames[0])
```

```python
filenames[-1]
```

```python
final_df = pd.DataFrame()
counter = 1
for file in tqdm(filenames[:-1]):
    with open(file, 'r') as f:
        df = pd.json_normalize(safe_load(f))
        df['match_id'] = counter
        final_df = final_df.append(df)
        counter+=1
        
final_df
```
