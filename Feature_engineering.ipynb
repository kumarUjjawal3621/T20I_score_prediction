### We observed that each element in 1st column contains ball-by-ball info of both the inning, and it is very compact, nested dictionaries and lists


### Final model input:  
#### BattingTeam, BowlingTeam, Venue, Inning, Overs(more than 10), Score, Wickets, Runs in last 5 overs, Target(if 2nd inning)
### Output : Final score after 20 overs

```python
df1.iloc[3,0][1]['2nd innings']['deliveries']
```

```python
dataset=[]

for i in range(len(df1)):
    try:
        dataset.append([])

        teams=df1.loc[i]['info.teams']
        toss_winner=df1.loc[i]['info.toss.winner']
        toss_choice=df1.loc[i]['info.toss.decision']
        batting_1st_inning = toss_winner if toss_choice == 'bat' else (set(teams) - set([toss_winner])).pop()
        bowling_1st_inning=(set(teams) - set([batting_1st_inning])).pop()
        # Ist inning
        df2=pd.DataFrame()
        score=0
        wickets=0
        deliveries=df1.iloc[i,0][0]['1st innings']['deliveries']
        for j in range(len(deliveries)):
            over=list(deliveries[j].keys())[0]
            run=deliveries[j][over]['runs']['total']
            score+=run
            wickets+=1 if 'wicket' in deliveries[j][over] else 0
            dic=pd.DataFrame({'over':[over],'score':[score],'wickets':[wickets]})
            df2=pd.concat([df2,dic],ignore_index=True)

        df2['batting']=batting_1st_inning
        df2['bowling']=bowling_1st_inning
        df2['inning']=1

        dataset[-1].append(df2)

        # IInd inning
        df2=pd.DataFrame()
        score=0
        wickets=0
        deliveries=df1.iloc[i,0][1]['2nd innings']['deliveries']
        for j in range(len(deliveries)):
            over=list(deliveries[j].keys())[0]
            run=deliveries[j][over]['runs']['total']
            score+=run
            wickets+=1 if 'wicket' in deliveries[j][over] else 0
            dic=pd.DataFrame({'over':[over],'score':[score],'wickets':[wickets]})
            df2=pd.concat([df2,dic],ignore_index=True)

        df2['batting']=bowling_1st_inning
        df2['bowling']=batting_1st_inning  
        df2['inning']=2
        dataset[-1].append(df2)
    except:
        print(i)
        dataset.pop()
```

#### We have successfully expanded the first column, now we have only two column left-
#### 'runs in last 5 overs','final_score'  --------- venue and target(if its 2nd inning) will not be doing

```python
for i in range(len(dataset)):
    print(i)
    for j in range(2):    ## j=0 or 1
        dataset[i][j]['Last5oversRuns']=-1
        for k in range(len(dataset[i][j])):
            over=dataset[i][j].loc[k]['over']
            if over>9.6:
                dataset[i][j].iloc[k,-1]=dataset[i][j].iloc[k,1] - dataset[i][j].iloc[k-30,1]
        dataset[i][j]['Final_score']=dataset[i][j].iloc[-1,1]
```
