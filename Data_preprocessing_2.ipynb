```python
# Count the frequencies of 'outcome' column
winning_counts = df['info.outcome.winner'].value_counts()

# Remove rows with frequencies less than 15
df = df[df['info.outcome.winner'].map(winning_counts) >= 15]
```

```python
no_of_matches=len(df)
print(no_of_matches)
df.iloc[0,0]
```

```python
column_list=['innings', 'info.dates', 'info.outcome.winner',
       'info.player_of_match', 'info.teams', 'info.toss.decision',
       'info.toss.winner', 'info.umpires', 'info.venue']
```

```python
def make_inning_df(inning_value,j):
    single_match=df.iloc[j,0]   
    balls_total=0  # total balls thrown in an inning, initialised
    deliveries_list=[]
    if inning_value==1:
        deliveries_list=single_match[0]['1st innings']['deliveries']
        balls_total=len(single_match[0]['1st innings']['deliveries'])
    else:
        deliveries_list=single_match[1]['2nd innings']['deliveries']
        balls_total=len(single_match[1]['2nd innings']['deliveries'])
        
    wickets_lost=0
    team_runs=0
    single_match_df=pd.DataFrame()
    for i in range(balls_total):
        over=list(deliveries_list[i].keys())[0]  # 0.1
        batsman=deliveries_list[i][over]['batsman']
        bowler=deliveries_list[i][over]['bowler']
        non_striker=deliveries_list[i][over]['non_striker']
        runs=deliveries_list[i][over]['runs']
        batsman_runs=runs['batsman']
        extra_runs=runs['extras']
        total_runs=runs['total']
        team_runs+=total_runs
        if 'extras' in deliveries_list[i][over]:
            extras=deliveries_list[i][over]['extras']
        else:
            extras='NaN'
        if 'wicket' in deliveries_list[i][over]:
            wicket=deliveries_list[i][over]['wicket']
            wickets_lost+=1
        else:
            wicket='NaN'

        new_row={
            'over':[over],'batsman':[batsman],'non_striker':[non_striker], 'bowler':[bowler],
            'batsman_runs':[batsman_runs], 'extra_runs':[extra_runs],'total_runs':[total_runs],
            'extras':[extras], 'wicket':[wicket], 'wickets_lost':[wickets_lost],'team_runs':[team_runs]
        }

        new_row=pd.DataFrame(new_row)
        single_match_df=pd.concat([single_match_df,new_row],ignore_index=True)
    return single_match_df
    
    
dataset=[]
for j in range(no_of_matches): 
    dataset.append([])    # each element is [ [match_info] ,{inning_1_df} , {inning_2_df} ]

    match_info={}   
    for k in range(1,len(column_list)):
        key=column_list[k]
        key=key.split('.',1)[1]
        match_info[key]=[df.iloc[j,k]]
    match_info=pd.DataFrame(match_info)
    dataset[j].append(match_info)
    
    inning_1_df=make_inning_df(1,j)
    dataset[j].append(inning_1_df)
    
    inning_2_df=make_inning_df(2,j)
    dataset[j].append(inning_2_df)
    print(j)
```
